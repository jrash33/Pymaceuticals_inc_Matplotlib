While your data companions rushed off to jobs in finance and government, you remained adamant that science was the way for you. Staying true to your mission, you've since joined Pymaceuticals Inc., a burgeoning pharmaceutical company based out of San Diego, CA. Pymaceuticals specializes in drug-based, anti-cancer pharmaceuticals. In their most recent efforts, they've since begun screening for potential treatments to squamous cell carcinoma (SCC), a commonly occurring form of skin cancer.

As their Chief Data Analyst, you've been given access to the complete data from their most recent animal study. In this study, 250 mice were treated through a variety of drug regimes over the course of 45 days. Their physiological responses were then monitored over the course of that time. Your objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

```python
 # Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
data_mouse = pd.read_csv(mouse_drug_data_to_load)
data_trial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data = pd.merge(data_trial, data_mouse, on="Mouse ID")

# Display the data table for preview
data.head()
```

### Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_vol = data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean().reset_index()

# Convert to DataFrame
mean_tumor_vol = pd.DataFrame(mean_tumor_vol)

# Preview DataFrame
mean_tumor_vol.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_vol = data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem().reset_index()

# Convert to DataFrame
sem_tumor_vol = pd.DataFrame(sem_tumor_vol)

# Preview DataFrame
sem_tumor_vol.head()
```

```python
 # Minor Data Munging to Re-Format the Data Frames
reformat_mean = mean_tumor_vol.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
reformat_sem = sem_tumor_vol.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
reformat_mean.head()
```

```python
# Generate the Plot (with Error Bars)

#Capomulin
plt.plot(reformat_mean.index,reformat_mean['Capomulin'], color='red', marker='o')

#Infubinol
plt.plot(reformat_mean.index,reformat_mean['Infubinol'], color='blue', marker='^')

#Ketapril
plt.plot(reformat_mean.index,reformat_mean['Ketapril'], color='green', marker='s')

#Placebo
plt.plot(reformat_mean.index,reformat_mean['Placebo'], color='black', marker='d')

#plot settings
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend()
plt.grid()

#Error bars
plt.errorbar(reformat_mean.index,reformat_mean['Capomulin'],reformat_sem['Capomulin'], color="red" )
plt.errorbar(reformat_mean.index,reformat_mean['Infubinol'],reformat_sem['Capomulin'], color="blue" )
plt.errorbar(reformat_mean.index,reformat_mean['Ketapril'],reformat_sem['Capomulin'], color="green" )
plt.errorbar(reformat_mean.index,reformat_mean['Placebo'],reformat_sem['Capomulin'], color="black" )

# Save the Figure
plt.savefig("Images/Tumor_Response_to_Treatment.png")

#show figure
plt.show()
```

### Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_meta = data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean().reset_index()

# Convert to DataFrame
mean_meta = pd.DataFrame(mean_meta)

# Preview DataFrame
mean_meta.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
sem_meta = data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem().reset_index()

# Convert to DataFrame
sem_meta = pd.DataFrame(sem_meta)

# Preview DataFrame
sem_meta.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_mean_meta = mean_meta.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
reformat_sem_meta = sem_meta.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
reformat_mean_meta.head()
```

```python
# Generate the Plot (with Error Bars)

#Capomulin
plt.plot(reformat_mean_meta.index,reformat_mean_meta['Capomulin'], color='red', marker='o')

#Infubinol
plt.plot(reformat_mean_meta.index,reformat_mean_meta['Infubinol'], color='blue', marker='^')

#Ketapril
plt.plot(reformat_mean_meta.index,reformat_mean_meta['Ketapril'], color='green', marker='s')

#Placebo
plt.plot(reformat_mean_meta.index,reformat_mean_meta['Placebo'], color='black', marker='d')

#plot settings
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (days)")
plt.ylabel("Met. Sites")
plt.legend()
plt.grid()

#Error bars
plt.errorbar(reformat_mean_meta.index,reformat_mean_meta['Capomulin'],reformat_sem_meta['Capomulin'], color="red" )
plt.errorbar(reformat_mean_meta.index,reformat_mean_meta['Infubinol'],reformat_sem_meta['Capomulin'], color="blue" )
plt.errorbar(reformat_mean_meta.index,reformat_mean_meta['Ketapril'],reformat_sem_meta['Capomulin'], color="green" )
plt.errorbar(reformat_mean_meta.index,reformat_mean_meta['Placebo'],reformat_sem_meta['Capomulin'], color="black" )

# Save the Figure
plt.savefig("Images/Metastatic_spread_during_treatment.png")

#show figure
plt.show()
```

### Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice = data.groupby(['Drug', 'Timepoint'])['Mouse ID'].count().reset_index()

# Convert to DataFrame
count_mice = pd.DataFrame(count_mice)

# Preview DataFrame
count_mice.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_count_mice = count_mice.pivot(index='Timepoint', columns='Drug', values='Mouse ID')

# Preview the Data Frame
reformat_count_mice.head()
```

```python
#Capomulin
plt.plot(reformat_count_mice.index,reformat_count_mice['Capomulin']*4, color='red', marker='o')

#Infubinol
plt.plot(reformat_count_mice.index,reformat_count_mice['Infubinol']*4, color='blue', marker='^')

#Ketapril
plt.plot(reformat_count_mice.index,reformat_count_mice['Ketapril']*4, color='green', marker='s')

#Placebo
plt.plot(reformat_count_mice.index,reformat_count_mice['Placebo']*4, color='black', marker='d')

#plot settings
plt.title("Survival During Treatment")
plt.xlabel("Time (days)")
plt.ylabel("Survival Rate (%)")
plt.legend()
plt.grid()

# Save the Figure
plt.savefig("Images/Survival_during_treatment.png")

#show figure
plt.show()
```

### Summary Bar Graph

```python
# Calculate the percent changes for each drug
percentage_change = (reformat_mean.iloc[-1,:]-reformat_mean.iloc[0,:])/reformat_mean.iloc[0,:]*100

# Display the data to confirm
percentage_change
```

# Store all Relevant Percent Changes into a Tuple

```python
#percentage_change = tuple(percentage_change)

#drugs we will plot
drugs = ['Capomulin','Infubinol','Ketapril','Placebo']
all_drugs = percentage_change[['Capomulin','Infubinol','Ketapril','Placebo']]

#PASS
pass_drugs = percentage_change['Capomulin']
pass_x_axis = ['Capomulin']

#FAIL
fail_drugs = percentage_change[['Infubinol','Ketapril','Placebo']]
fail_x_axis =['Infubinol','Ketapril','Placebo']

# Create the ticks for our bar chart's x axis
x_tick_locations=np.arange(len(drugs))+.5
plt.xticks(x_tick_locations, drugs)

# Create the ticks for our bar chart's y axis
y_tick_locations=[-20, 0, 20, 40, 60]
plt.yticks(y_tick_locations, [-20, 0, 20, 40, 60])

#plot
plt.bar(pass_x_axis, pass_drugs, color='g', align="center", width=1)
plt.bar(fail_x_axis,fail_drugs, color='r', align="center", width=1)

#value labels on bars
plt.text(-.2,-10, f"{int(all_drugs[0])}%", color='w')
plt.text(.9, 10, f"{int(all_drugs[1])}%", color='w')
plt.text(1.9, 10, f"{int(all_drugs[2])}%", color='w')
plt.text(2.9, 10, f"{int(all_drugs[3])}%", color='w')

#plot settings
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")
plt.ylim(-30,70)
plt.grid()

# Save the Figure
plt.savefig("Images/Tumor_change_over_45_days.png")

plt.show()
```
